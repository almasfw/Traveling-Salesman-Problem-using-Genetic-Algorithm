```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def createPopulation(cities, populationSize):
    population = []
    initialPoint = np.array([1])
    for i in range(populationSize):
        temp = np.array(random.sample(cities, len(cities)))
        temp = np.delete(temp, np.where(temp == 1))
        population.append(np.concatenate([initialPoint, temp]))
    
    return np.array(population)
```

```python
def calculateDistanceFitness(dMatrix, population):
    distance = []
    fitness = []
    nCity = dMatrix.shape[0]
    d = 0
    for route in population:
        for i in range(nCity):
            city1 = route[i]
            city2 = route[(i+1) % nCity]
            d += dMatrix[int(city1)-1][int(city2)-1]
        distance.append(d)
        fitness.append(1/d)
        d = 0
    
    distance = np.array(distance)
    fitness = np.array(fitness)
    
    sortedDistance = np.sort(distance)
    sortedFitness = -(np.sort(-1*fitness))
    sortIndex = distance.argsort()
    sortedPopulation = population[sortIndex]
    
    return sortedPopulation, sortedDistance, sortedFitness
```

```python
def matingPool(population, fitness, eliteSize):
    mates = []
    nPopulation = population.shape[0]
    cum_sum = np.cumsum(fitness)/np.sum(fitness)
    for i in range(eliteSize):
        mates.append(population[i])
    for i in range(nPopulation - eliteSize):
        pick = random.random()
        for j in range(nPopulation):
            if pick <= cum_sum[j]:
                mates.append(population[i])
                break
    
    return np.array(mates)
```

```python
def breeding(parents1, parents2):
    nGen = parents1.shape[0]
    index1 = random.randint(0, nGen-1)
    index2 = random.randint(0, nGen-1)
    child = np.zeros(nGen)
    
    start = min(index1, index2)
    end = max(index1, index2)
    
    for i in range(start, end+1):
        child[i] = parents1[i]
    
    for gen in parents2:
        for i in range(0, start):
            if not np.isin(gen, child):
                if child[i] == 0:
                    child[i] = gen
        for i in range(end+1, nGen):
            if not np.isin(gen, child):
                if child[i] == 0:
                    child[i] = gen
    
    return np.array(child)
    
def breedingPopulation(parents, eliteSize):
    children = []
    nPopulation = len(parents)
    nChildren = nPopulation - eliteSize
    
    for i in range(eliteSize):
        children.append(parents[i])
    
    for i in range(nChildren):
        parentsIndex = np.random.randint(low = len(parents), size = 2)
        parents1 = parents[parentsIndex[0]]
        parents2 = parents[parentsIndex[1]]
        child = breeding(parents1, parents2)
        children.append(child)
    
    return np.array(children)
```

```python
def mutate(individu):
    nGen = len(individu)
    indexSwap = np.random.randint(1, nGen, size = 2)
    swap1 = indexSwap[0]
    swap2 = indexSwap[1]
    temp = individu[swap1]
    individu[swap1] = individu[swap2]
    individu[swap2] = temp
    
    return individu

def mutatePopulation(population, mutationRate):
    mutated = []
    
    for ind in population:
        if(random.random()<mutationRate):
            mutated.append(mutate(ind))
        else:
            mutated.append(ind)
    
    return np.array(mutated)
```

```python
def newGeneration(dMatrix, currentGeneration, eliteSize, mutationRate):
    #calculate distance, fitness, and sort the population according to fitness/distance
    population, distance, fitness = calculateDistanceFitness(dMatrix, currentGeneration)
    bestFitness = fitness[0]
    
    parents = matingPool(population, fitness, eliteSize)
    children = breedingPopulation(parents, eliteSize)
    nextGeneration = mutatePopulation(children, mutationRate)
    
    return nextGeneration, bestFitness
```

```python
def geneticAlgorithm(dMatrix, cities, populationSize, eliteSize, mutationRate, generation):
    #initialize population
    population = createPopulation(cities, populationSize)
    progress = []
    
    for i in range(generation):
        population, bestFitness = newGeneration(dMatrix, population, eliteSize, mutationRate)
        progress.append(bestFitness)
    
    sortedPopulation, distance, fitness = calculateDistanceFitness(dMatrix, population)
    bestRoute = sortedPopulation[0]
    
    return bestRoute, progress
```

```python
def plotRoute(cities, bestRoute):
    x = cities[1][bestRoute-1]
    y = cities[2][bestRoute-1]
    
    figure_size = 15
    plt.plot(x, y, 'ro-', ms=1, mec='k')
    plt.title('Best Route'), plt.xlabel('x'), plt.ylabel('y')
    plt.show()
```

```python
#import data

# dMatrix1 = pd.read_excel (r'yk11.xls', sheet_name='DistanceMatrix', header=None)
# dMatrix1 = np.array(dMatrix1)

cities1 = pd.read_excel(r'yk11.xls', sheet_name='Data', header=None)
cityName1 = set(cities1[0])
nCity1 = len(cities1)
dMatrix1 = np.zeros([nCity1, nCity1])
for i in range(nCity1):
    for j in range(i+1):
        d = ((cities1[1][i] - cities1[1][j])**2 + (cities1[2][i] - cities1[2][j])**2) ** 1/2
        dMatrix1[i][j] = d
        dMatrix1[j][i] = d

populationSize = 100
eliteSize = 20
mutationRate = 0.1
generation = 200

bestRoute1, progress1 = geneticAlgorithm(dMatrix1, cityName1, populationSize, eliteSize, mutationRate, generation)
bestRoute1 = np.concatenate([bestRoute1, np.array([1])])
print(bestRoute1)
```

```python
iteration = [i+1 for i in range(generation)]

plt.plot(iteration, progress1, 'ro-', ms=1, mec='k')
plt.title('Progress Case 1'), plt.xlabel('Generation'), plt.ylabel('Progress')
plt.show()

plotRoute(cities1, bestRoute1)
```
